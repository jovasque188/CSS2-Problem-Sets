# Block 1

```python
# Import libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Block 2

```python
# Load data 
data = sns.load_dataset('anscombe')
```

```python
# plot the figures 

x1 = data.x[0:11]
x2 = data.x[11:22]
x3 = data.x[22:33]
x4 = data.x[33:]
y1 = data.y[0:11]
y2 = data.y[11:22]
y3 = data.y[22:33]
y4 = data.y[33:]

fig, axes = plt.subplots(2,2,figsize=[10,10])

# Create subplot I
axes[0,0].plot(x1,y1, linestyle='', marker='*', color='r') 
axes[0,0].set_title('I')
axes[0,0].set_ylabel('y1')
axes[0,0].set_xlabel('x1')

# Create subplot II
axes[0,1].plot(x2,y2, linestyle='', marker='*', color='y') 
axes[0,1].set_title('II')
axes[0,1].set_xlabel('x2')
axes[0,1].set_ylabel('y2')

# Create subplot III
axes[1,0].plot(x3,y3, linestyle='', marker='*', color='b') 
axes[1,0].set_title('III')
axes[1,0].set_xlabel('x3')
axes[1,0].set_ylabel('y3')

# Create subplot IV
axes[1,1].plot(x4,y4, linestyle='', marker='*', color='k') 
axes[1,1].set_title('IV')
axes[1,1].set_ylabel('y4')
axes[1,1].set_xlabel('x4')

# Add texts to subplots with mean and std of x and y
x1mean = np.mean(data.x)
x1std = round(np.std(data.x), 2)
y1mean = round(np.mean(data.y), 1)
y1std = round(np.std(data.y), 2)
axes[0,0].text(13, 1, f'x mean: {x1mean}\ny mean: {y1mean}\nx std: {x1std}\ny std: {y1std}', fontsize=12)
axes[0,1].text(13, 1, f'x mean: {x1mean}\ny mean: {y1mean}\nx std: {x1std}\ny std: {y1std}', fontsize=12)
axes[1,0].text(13, 1, f'x mean: {x1mean}\ny mean: {y1mean}\nx std: {x1std}\ny std: {y1std}', fontsize=12)
axes[1,1].text(13, 1, f'x mean: {x1mean}\ny mean: {y1mean}\nx std: {x1std}\ny std: {y1std}', fontsize=12)

# Set limits to be the same for all subplots 
for i in np.arange(0, axes.shape[0]):
  for j in np.arange(0,axes.shape[1]):
    axes[i,j].set_ylim([0,14])
    axes[i,j].set_xlim([0,20])
```
